```python
import ast
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/LLM ASSIGNMENT/cocktails_recipe.csv')

# Parse ingredients from stringified lists to actual lists
df['parsed_ingredients'] = df['ingredients'].apply(lambda x: [item[1] for item in ast.literal_eval(x)])

# Create tags manually based on common categories (you can expand this list)
tags = ['vodka', 'gin', 'rum', 'tequila', 'sweet', 'sour', 'whiskey', 'brandy', 'lime', 'lemon', 'orange']

# Create a function to assign tags
def extract_tags(ingredient_list):
    tags_found = []
    for ing in ingredient_list:
        for tag in tags:
            if tag.lower() in ing.lower():
                tags_found.append(tag)
    return list(set(tags_found))

# Apply tagging
df['tags'] = df['parsed_ingredients'].apply(extract_tags)

# Remove empty tags
df = df[df['tags'].map(len) > 0].reset_index(drop=True)

# Binarize the tags for multi-label classification
mlb = MultiLabelBinarizer()
tag_matrix = mlb.fit_transform(df['tags'])

# Add a column with full ingredients as a string for BERT input
df['text'] = df['parsed_ingredients'].apply(lambda x: ' '.join(x))
```

```python
from datasets import Dataset
import torch

# Create HuggingFace dataset
hf_dataset = Dataset.from_pandas(df[['text', 'tags']])
hf_dataset = hf_dataset.train_test_split(test_size=0.1)

# Add multi-hot encoded labels
hf_dataset = hf_dataset.map(lambda x: {
    'labels': mlb.transform([x['tags']])[0]
})
```

```python
!pip install -U transformers
from transformers import TrainingArguments, Trainer
from sklearn.metrics import f1_score, precision_score, recall_score
import torch
import numpy as np

# Optional: define collate_fn if you're not using it elsewhere
def collate_fn(batch):
    return {
        'input_ids': torch.stack([x['input_ids'] for x in batch]),
        'attention_mask': torch.stack([x['attention_mask'] for x in batch]),
        'labels': torch.tensor([x['labels'] for x in batch], dtype=torch.float)
    }

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    evaluation_strategy="epoch",  # evaluate at end of each epoch
    logging_dir="./logs",
    logging_steps=10,
    report_to="none"
)

# Evaluation metrics
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    probs = torch.sigmoid(torch.tensor(logits)).numpy()
    preds = (probs >= 0.5).astype(int)
    return {
        'f1': f1_score(labels, preds, average='micro', zero_division=0),
        'precision': precision_score(labels, preds, average='micro', zero_division=0),
        'recall': recall_score(labels, preds, average='micro', zero_division=0)
    }

# Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=hf_dataset["train"],
    eval_dataset=hf_dataset["test"],
    tokenizer=tokenizer,
    data_collator=collate_fn,
    compute_metrics=compute_metrics
)

# Train
trainer.train()

# Final evaluation
metrics = trainer.evaluate()
print("\n Evaluation Metrics:")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}")

# Loss visualization
import matplotlib.pyplot as plt

logs = trainer.state.log_history
loss = [entry["loss"] for entry in logs if "loss" in entry]
steps = [entry["step"] for entry in logs if "loss" in entry]

plt.figure(figsize=(8, 5))
plt.plot(steps, loss, marker='o', color='teal')
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Install necessary libraries
!pip install transformers datasets evaluate scikit-learn --quiet
!pip install --upgrade transformers --quiet

# ----------------------------------------
# Step 1: Load and preprocess the dataset
# ----------------------------------------

import ast
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Load the CSV dataset
df = pd.read_csv('/content/drive/MyDrive/LLM ASSIGNMENT/cocktails_recipe.csv')

# Parse ingredient tuples and extract ingredient names
df['parsed_ingredients'] = df['ingredients'].apply(lambda x: [item[1] for item in ast.literal_eval(x)])

# Define tags (can be expanded)
tags = ['vodka', 'gin', 'rum', 'tequila', 'sweet', 'sour', 'whiskey', 'brandy', 'lime', 'lemon', 'orange']

# Extract tags from ingredients
def extract_tags(ingredient_list):
    return list({tag for tag in tags if any(tag in ing.lower() for ing in ingredient_list)})

df['tags'] = df['parsed_ingredients'].apply(extract_tags)
df = df[df['tags'].map(len) > 0].reset_index(drop=True)

# Binarize tags for multi-label classification
mlb = MultiLabelBinarizer()
tag_matrix = mlb.fit_transform(df['tags'])

# Create a single string for BERT input
df['text'] = df['parsed_ingredients'].apply(lambda x: ' '.join(x))


# ----------------------------------------
# Step 2: Convert to Hugging Face Dataset
# ----------------------------------------

from datasets import Dataset
import torch

# Wrap into Hugging Face dataset and split
hf_dataset = Dataset.from_pandas(df[['text', 'tags']])
hf_dataset = hf_dataset.train_test_split(test_size=0.1)

# Add multi-hot labels
hf_dataset = hf_dataset.map(lambda x: {
    'labels': mlb.transform([x['tags']])[0]
})


# ----------------------------------------
# Step 3: Tokenization
# ----------------------------------------

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True)

hf_dataset = hf_dataset.map(tokenize, batched=True)
hf_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])


# ----------------------------------------
# Step 4: Load model
# ----------------------------------------

from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=len(mlb.classes_),
    problem_type="multi_label_classification"
)


# ----------------------------------------
# Step 5: Define training args
# ----------------------------------------

from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    logging_dir="./logs",
    logging_steps=10,
    report_to="none",
    logging_strategy="steps",
    save_strategy="no"
)


# ----------------------------------------
# Step 6: Define evaluation metrics
# ----------------------------------------

from sklearn.metrics import f1_score, precision_score, recall_score
import numpy as np

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    probs = torch.sigmoid(torch.tensor(logits)).numpy()
    preds = (probs >= 0.5).astype(int)
    labels = np.array(labels)
    return {
        'f1': f1_score(labels, preds, average='micro'),
        'precision': precision_score(labels, preds, average='micro'),
        'recall': recall_score(labels, preds, average='micro')
    }


# ----------------------------------------
# Step 7: Training
# ----------------------------------------

from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=hf_dataset["train"],
    eval_dataset=hf_dataset["test"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()


# ----------------------------------------
# Step 8: Evaluation
# ----------------------------------------

metrics = trainer.evaluate()
print("Evaluation metrics:", metrics)


# ----------------------------------------
# Step 9: Visualization
# ----------------------------------------

import matplotlib.pyplot as plt

logs = trainer.state.log_history
loss = [entry["loss"] for entry in logs if "loss" in entry]
steps = [entry["step"] for entry in logs if "loss" in entry]

plt.plot(steps, loss, marker='o')
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.show()


# ----------------------------------------
# Step 10: Test a sample prediction
# ----------------------------------------

sample = df.iloc[0]
inputs = tokenizer(sample["text"], return_tensors="pt", truncation=True, padding=True)
inputs = {k: v.to(model.device) for k, v in inputs.items()}

outputs = model(**inputs)
probs = torch.sigmoid(outputs.logits).detach().cpu().numpy()[0]
predicted_tags = [mlb.classes_[i] for i, p in enumerate(probs) if p > 0.5]

print("Cocktail:", sample["title"] if "title" in sample else "N/A")
print("Ingredients:", sample["text"])
print("True Tags:", sample["tags"])
print("Predicted Tags:", predicted_tags)
```
